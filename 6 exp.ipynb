```python
df['10DF']=df['Variable'].diff()
df['20DF']=df['10DF'].diff()
df['30DF']=df['20DF'].diff()
```

```python
plt.plot(df.index,df['Variable'],color='brown')
plt.plot(df.index,df['10DF'],color='blue')
plt.plot(df.index,df['20DF'],color='red')
plt.plot(df.index,df['30DF'],color='pink')
plt.legend(['Variable','10DF','20DF','30DF'])
plt.xlabel('Timestamp')
plt.ylabel('Variable')
plt.show()
```

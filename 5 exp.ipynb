```python
time_series_variable = 'R1 (MOhm)'

fig, axes = plt.subplots(2, 1, figsize=(10, 8))

# Plot ACF (Autocorrelation Function)
plot_acf(df[time_series_variable], ax=axes[0], lags=40)
axes[0].set_title(f'ACF of {time_series_variable}')

# Plot PACF (Partial Autocorrelation Function)
plot_pacf(df[time_series_variable], ax=axes[1], lags=40)
axes[1].set_title(f'PACF of {time_series_variable}')

plt.show();
```

(b) Perform ADF test to check the given TS variable for stationarity.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
time_series_variable = 'R1 (MOhm)'
result = adfuller(df[time_series_variable], autolag='AIC')

# Print the ADF test results
print(f'ADF Statistic: {result[0]}')
print(f'p-value: {result[1]}')
print('Critical Values:')
for key, value in result[4].items():
    print(f'{key}: {value}')
    

if result[1] <= 0.05:
    print("Reject the null hypothesis: The time series is likely stationary.")
else:
    print("Fail to reject the null hypothesis: The time series is likely non-stationary.")
```

(c) Write a python function/method that returns the ARIMA order for the given TS variable.

```python
from pmdarima import auto_arima
```

```python
def get_arima_order(time_series_data):
    
    arima_model = auto_arima(time_series_data, seasonal=False, trace=True, error_action='ignore', suppress_warnings=True)
    
    arima_order = arima_model.order
    return arima_order
```

```python
time_series_variable = 'R1 (MOhm)'
arima_order = get_arima_order(df[time_series_variable])

print(f"ARIMA Order for {time_series_variable}: {arima_order}")
```

(d) Forecast the TS variables for the following horizons (h=5, 10,15, and 50) and compute Mean
Absolute Error (MAE).

```python
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_absolute_error
```

```python

model = ARIMA(df["R1 (MOhm)"], order=(1, 1, 1))
model_fit = model.fit()

# Horizons for forecasting
horizons = [5, 10, 15, 50]

for h in horizons:
    # Forecast the time series variable
    forecast = model_fit.forecast(steps=h)

    # Calculate Mean Absolute Error (MAE)
    actual_values = df["R1 (MOhm)"].iloc[-h:]
    mae = mean_absolute_error(actual_values, forecast)

    print(f"Forecast for horizon {h}: {forecast}")
    print(f"MAE for horizon {h}: {mae}\n")
```

(e) Plot actual and forecasted values of the TS variable.

```python
# Read data from CSV file
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv(r'C:\Users\Suyash Tambe\Desktop\PDS Assignment 2\gas+sensor+array+temperature+modulation\gas-sensor-array-temperature-modulation\20160930_203718.csv')

horizons = [5, 10, 15, 50]

model = ARIMA(df["R1 (MOhm)"], order=(1, 1, 1))
model_fit = model.fit()
```

```python
for h in horizons:
    # Forecast the time series variable
    forecast = model_fit.forecast(steps=h)
    
    # Calculate Mean Absolute Error (MAE)
    actual_values = df["R1 (MOhm)"].iloc[-h:]
    mae = mean_absolute_error(actual_values, forecast)
    
    print(f"Forecast for horizon {h}: {forecast}")
    print(f"MAE for horizon {h}: {mae}\n")
    
    # Plot actual and forecasted values
    plt.figure()
    plt.plot(df.index, df["R1 (MOhm)"], label="Actual")
    plt.plot(pd.DataFrame(df.index).iloc[-h:], forecast, label="Forecast")
    plt.xlabel("Time")
    plt.ylabel("R1 (MOhm)")
    plt.title(f"Actual vs Forecast for Horizon {h}")
    plt.legend()
    plt.show()
```
